```python
def mini(imag):
    imagaux=imag
    miniatura = (160, 120)
    imagaux.thumbnail(miniatura)
    return(imagaux)
```

```python
au=mini(imagen)
```

```python
au.show()
```

```python
def SimHo(imag):
    mp=imag.load()
    imagaux=Image.new("RGB",(imag.size[0]*2,imag.size[1]),"white")
    mpaux=imagaux.load()
    for i in range(imag.size[0]):
        for j in range(imag.size[1]):
            mpaux[i,j]=mp[i,j]
    for i in range(imag.size[0]):
        for j in range(imag.size[1]):
            mpaux[imag.size[0]+i,j]=mpaux[imag.size[0]-i-1,j]
    return(imagaux)
```

```python
au2=SimHo(au)
```

```python
au2.show()
```

```python
def SimVe(imag):
    mp=imag.load()
    imagaux=Image.new("RGB",(imag.size[0],imag.size[1]*2),"white")
    mpaux=imagaux.load()
    for i in range(imag.size[0]):
        for j in range(imag.size[1]):
            mpaux[i,j]=mp[i,j]
    for i in range(imag.size[0]):
        for j in range(imag.size[1]):
            mpaux[i,imag.size[1]+j]=mpaux[i,imag.size[1]-j-1]
    return(imagaux)
```

```python
au3=SimVe(au)
```

```python
au3.show()
```

```python
def pmm(imag,rang):
    a3=imag
    for i in range(rang):
        a1=SimHo(a3)
        a2=SimVe(a1)
        a3=a2
        ImageSequence.Iterator(a3)
        i=i+1
    return(a3)
```

```python
pmm(au,2)
```

```python
def Matrotacion(angulo):
    M=array([[math.cos(angulo),-math.sin(angulo)],[math.sin(angulo),math.cos(angulo)]])
    return(M)
```

```python
Matrotacion(math.pi)
```

```python
def Rotacion(imag,angulo):
    mp=imag.load()
    imagaux=Image.new("RGB",(imag.size[0]*10,imag.size[1]*10),"white")
    mpaux=imagaux.load()
    M=Matrotacion(angulo)
    for i in range(imag.size[0]):
        for j in range(imag.size[1]):
            mpaux[i,j]=mp[i,j]
    for i in range(imag.size[0]):
        for j in range(imag.size[1]):
            mpaux[i,imag.size[1]+j]=mpaux[floor(np.dot(M,[i,j]).tolist()[0])+imag.size[0],floor(np.dot(M,[i,j]).tolist()[1])+imag.size[1]]
    return(imagaux)
```

```python
Rotacion(au,math.pi)
```
