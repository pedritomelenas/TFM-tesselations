## Parrilla

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%pylab inline
```

La función parrilla creará una parrilla con variables:
- pendiente = pendiente de las rectas que formarán la parrilla
- anchura = para fijar la distancia entre las rejillas, debe ser apropiada para que cubra toda la región
- xylim = para fijar los límites de los ejes de coordenadas
- transparencia = grado de transparencia de la parrila entre 0 y 1

```python
def parrilla(pendiente,anchura,xylim,transparencia):
    x=np.arange(-xylim,xylim)
    y=pendiente*x
    if pendiente==0: 
        for index in range(-xylim/anchura,xylim/anchura):
            plt.plot([anchura*index,anchura*index],[-xylim,xylim],"b",alpha=transparencia)
            plt.plot([-xylim,xylim],[anchura*index,anchura*index],"b",alpha=transparencia)
    else:
        for index in range (-xylim,xylim):
            plt.plot(x,y+anchura*index,"b",alpha=transparencia)
            plt.plot(x,-y+anchura*index,"b",alpha=transparencia)
    plt.ylim(-xylim,xylim)
    plt.xlim(-xylim,xylim)
    plt.plot([0,0],[-xylim,xylim],"r")
    plt.plot([-xylim,xylim],[0,0],"r")
plt.show()
```

```python
parrilla(6,20,40,0.5)
```

```python
parrilla(6,20,40,0.4)
```

```python
parrilla(6,10,40,0.3)
```

Si la pendiente es cero

```python
parrilla(0,5,40,0.3)
```
